```python
'''
Used to generate missing fasta files that were not executed successfull from alphafold 2
'''

import os
import re
import shutil

def get_missing_files(directory_path, pattern, max_count=0):
    """
    Reads .zip files from directory, identifies missing indexes based on filename:
    ecoli_group_6_25_{count}.result.zip
    
    Returns a list of missing count values.
    """
    
    # Regex to extract numbers from filenames
    
    
    found_counts = set()
    
    # Loop through directory and extract counts
    for filename in os.listdir(directory_path):
        match = pattern.match(filename)
        if match:
            count = int(match.group(1))
            found_counts.add(count)
    
    # Determine missing counts
    missing = [i for i in range(1, max_count + 1) if i not in found_counts]
    
    return missing

def copy_selected_files(source_dir, filenames, save_dir):
    """
    Append directory path with each file name to create full paths.
    Copy each file to the target save directory.
    """

    # Ensure the save directory exists
    os.makedirs(save_dir, exist_ok=True)

    copied_files = []
    missing_files = []

    for fname in filenames:
        src_path = os.path.join(source_dir, fname)
        dst_path = os.path.join(save_dir, fname)

        # Check if file exists
        if os.path.exists(src_path):
            shutil.copy2(src_path, dst_path)   # copy2 keeps metadata
            copied_files.append(fname)
        else:
            missing_files.append(fname)

    return copied_files, missing_files
```

```python
#Stap missing files finder
pattern0 = re.compile(r"stapc_group_1_5_(\d+)\.result\.zip$")
pattern = re.compile(r"stapc_group_6_25_(\d+)\.result\.zip$")
pattern2 = re.compile(r"stapc_group_26_49_(\d+)\.result\.zip$")
pattern3 = re.compile(r"stapc_group_50_plus_(\d+)\.result\.zip$")

missing_files0 = get_missing_files("/data/prem001/PGAT-ABPp/code/data/alphafold_pdb_stap/result", pattern0, max_count=114)
missing_files = get_missing_files("/data/prem001/PGAT-ABPp/code/data/alphafold_pdb_stap/result", pattern, max_count=3387)
missing_files2 = get_missing_files("/data/prem001/PGAT-ABPp/code/data/alphafold_pdb_stap/result", pattern2, max_count=810)
missing_files3 = get_missing_files("/data/prem001/PGAT-ABPp/code/data/alphafold_pdb_stap/result", pattern3, max_count=70)
missing_file_names = [f"stapc_group_1_5_{index}.fasta" for index in missing_files0] + \
                     [f"stapc_group_6_25_{index}.fasta" for index in missing_files] + \
                     [f"stapc_group_26_49_{index}.fasta" for index in missing_files2] + \
                     [f"stapc_group_50_plus_{index}.fasta" for index in missing_files3]



source_dir = "/data/prem001/PGAT-ABPp/code/data/fasta_stapc/all"
save_to = "/data/prem001/PGAT-ABPp/code/data/fasta_stapc/all/miss"
copied, missing = copy_selected_files(source_dir, missing_file_names, save_to)
print("Copied:", copied)
print("Missing:", missing)
```
